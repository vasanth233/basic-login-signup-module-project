```python
import re
import getpass
import os

# --- STAGE 1: VALIDATION FUNCTIONS ---

def is_valid_username(username):
    """Validates the username (email) based on PDF criteria."""
    # Rule 3: Must start with a letter
    if not username or not username[0].isalpha():
        return False
    # Check for basic structure: '@' and '.' after '@', no immediate '.' after '@'
    parts = username.split('@')
    if len(parts) != 2 or '.' not in parts[1] or parts[1].startswith('.'):
        return False
    return True

def is_valid_password(password):
    """Validates the password based on PDF criteria."""
    # Length 6-16
    if not (6 <= len(password) <= 16): return False
    # Requires uppercase, lowercase, digit, and special character
    if not re.search(r'[A-Z]', password): return False
    if not re.search(r'[a-z]', password): return False
    if not re.search(r'[0-9]', password): return False
    if not re.search(r'[^a-zA-Z0-9\s]', password): return False
    return True

# --- STAGE 2: DATA HANDLING FUNCTIONS ---

def load_users():
    """Reads users.txt and returns a dictionary {username: password}."""
    users = {}
    if not os.path.exists("users.txt"):
        return users # Return empty dict if file doesn't exist

    try:
        with open("users.txt", "r") as f:
            for line in f:
                line = line.strip()
                if line:
                    username, password = line.split(':', 1)
                    users[username] = password
    except Exception as e:
        print(f"ðŸš¨ Error loading user data: {e}")
    return users

def store_credentials(username, password):
    """Stores the validated username and password in 'users.txt' using append mode ('a')."""
    try:
        # Use 'a' to append data without overwriting
        with open("users.txt", "a") as f:
            f.write(f"{username}:{password}\n")
        return True
    except IOError:
        print("ðŸš¨ Error: Could not write to the users file.")
        return False

def update_user_password(username, new_password, all_users):
    """Updates the password for an existing user and rewrites the file."""
    all_users[username] = new_password
    try:
        # Overwrite the file ('w') with the updated dictionary content
        with open("users.txt", "w") as f:
            for user, pwd in all_users.items():
                f.write(f"{user}:{pwd}\n")
        return True
    except IOError:
        print("ðŸš¨ Error: Could not update the users file.")
        return False

def register_user(all_users):
    """Handles user registration with validation loops and checks for existing users."""
    print("\n--- NEW USER REGISTRATION ---")
    while True:
        username = input("âž¡ï¸ Enter Username (Email): ").strip()
        if not is_valid_username(username):
            print("âŒ Invalid Username. Must start with a letter, contain '@', and a '.' in the domain.")
        elif username in all_users:
            print("âŒ Username already exists. Please login or choose another username.")
        else:
            break

    while True:
        password = getpass.getpass("âž¡ï¸ Enter Password: ")
        if is_valid_password(password):
            break
        else:
            print("âŒ Invalid Password. Must be 6-16 chars, contain U/L case, digit, and special char.")

    if store_credentials(username, password):
        print(f"\nâœ… User '{username}' successfully registered!")

# --- STAGE 3: LOGIN & FORGOT PASSWORD ---

def forgot_password_menu(all_users):
    """Handles the Forgot Password workflow as per PDF (retrieve/update)."""
    print("\n--- FORGOT PASSWORD ---")
    username = input("âž¡ï¸ Enter your Username (Email): ").strip()

    if username in all_users:
        while True:
            # Check if current password is known to allow update
            current_pwd_check = getpass.getpass("ðŸ” Enter your current password (or press Enter to see it): ")

            if not current_pwd_check:
                # Option 1: Display password (as per PDF requirement)
                print(f"\nðŸ”‘ Your retrieved password is: **{all_users[username]}**")
                return
            elif current_pwd_check == all_users[username]:
                # Option 2: Credentials match, allow setting a new password (as per PDF requirement)
                print("\nâœ… Current password verified. You can now set a new one.")
                while True:
                    new_password = getpass.getpass("âž¡ï¸ Enter NEW Password: ")
                    if is_valid_password(new_password):
                        if update_user_password(username, new_password, all_users):
                            print("âœ… Password successfully updated!")
                        return
                    else:
                        print("âŒ Invalid New Password. Please meet all criteria.")
            else:
                print("âŒ Incorrect current password.")
                return
    else:
        # Username not found (as per PDF requirement)
        print("âŒ Username not found. Please register a new account.")

def login_user(all_users):
    """Handles user login authentication."""
    print("\n--- USER LOGIN ---")
    username = input("âž¡ï¸ Enter Username (Email): ").strip()
    password = getpass.getpass("âž¡ï¸ Enter Password: ")

    if username in all_users and all_users[username] == password:
        print(f"\nðŸŽ‰ Welcome back, {username}! Login successful.")
        return True
    else:
        print("\nâŒ Invalid credentials.")

        while True:
            choice = input("Do you want to (R)egister, (F)orgot Password, or (T)ry again? ").upper()
            if choice == 'R':
                register_user(all_users)
                return False
            elif choice == 'F':
                forgot_password_menu(all_users)
                return False
            elif choice == 'T':
                return login_user(all_users) # Recursive call to try login again
            else:
                print("Invalid choice.")

# --- MAIN PROGRAM LOOP ---

def run_program():
    """The main entry point for the application."""
    while True:
        all_users = load_users()

        print("\n==================================")
        print("ðŸ‘¤ Login/Registration System Menu")
        print("==================================")
        print("1. Login")
        print("2. Register")
        print("3. Exit")

        choice = input("Enter your choice (1-3): ").strip()

        if choice == '1':
            login_user(all_users)
        elif choice == '2':
            register_user(all_users)
        elif choice == '3':
            print("Goodbye! ðŸ‘‹")
            break
        else:
            print("Invalid choice. Please enter 1, 2, or 3.")

if __name__ == "__main__":
    run_program()
```
